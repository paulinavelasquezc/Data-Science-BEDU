## Análisis Exploratorio de Datos

### Definición de columnas:

La siguiente información fue obtenida del manual de datos dado por Kaggle:

**work_year:** El año en que se registraron los datos. Este campo indica el contexto temporal de los datos, importante para comprender las tendencias salariales a lo largo del tiempo.

**job_title:** El título específico de la función laboral, como "Científico de datos", "Ingeniero de datos" o "Analista de datos". Esta columna es crucial para comprender la distribución salarial entre las distintas funciones especializadas dentro del campo de los datos.

**job_category:** Clasificación del puesto de trabajo en categorías más amplias para facilitar el análisis. Puede incluir áreas como "Análisis de datos", "Aprendizaje automático", "Ingeniería de datos", etc.

**salary_currency:** La moneda en la que se paga el salario, como USD, EUR, etc. Esto es importante para la conversión de divisas y para comprender el valor real del salario en un contexto global.

**salary:** El salario bruto anual del rol en la moneda local. Esta cifra de salario bruto es clave para las comparaciones salariales regionales directas.

**salary_in_usd:** El salario bruto anual convertido a dólares estadounidenses (USD). Esta conversión de moneda uniforme ayuda en las comparaciones y análisis salariales globales.

**employee_residence:** El país de residencia del empleado. Este dato puede utilizarse para explorar las diferencias salariales geográficas y las variaciones del coste de la vida.

**experience_level:** Clasifica el nivel de experiencia profesional del empleado. Las categorías más comunes pueden ser "principiante", "medio", "superior" y "ejecutivo", lo que permite comprender cómo influye la experiencia en el salario de los puestos relacionados con los datos.

**employment_type:** Especifica el tipo de empleo, como "jornada completa", "jornada parcial", "contrato", etc. Esto ayuda a analizar cómo afectan las diferentes modalidades de empleo a las estructuras salariales.

**work_setting:** El entorno de trabajo, como "A distancia", "En persona" o "Híbrido". Esta columna refleja el impacto de los entornos de trabajo en los niveles salariales del sector de datos.

**company_location:** El país en el que está ubicada la empresa. Ayuda a analizar cómo afecta la ubicación de la empresa a las estructuras salariales.

**company_size:**  El tamaño de la empresa empleadora, a menudo clasificada en tamaño pequeño (S), mediano (M) y grande (L). Permite analizar cómo influye el tamaño de la empresa en el salario.

## Estimados de locación y variabilidad

### Media, media truncada, mediana y moda

```python
promedio_formateado = df['salary_in_usd'].mean()
mediana_formateado = df['salary_in_usd'].median()
moda_formateado = df['salary_in_usd'].mode()[0]
```

```python
print("Media: ${:,.3f}, \nMediana: ${:,.3f}, \nModa: ${:,.3f}".format(promedio_formateado, mediana_formateado, moda_formateado))
```

```python
media_truncada_formateado = stats.trim_mean(df['salary_in_usd'], 0.1)
print(f'Media truncada: ${media_truncada_formateado:,.3f}')
```

### Desviación estándar

```python
desviacion_estandar = df['salary_in_usd'].std()
print("Desviación estándar: ${:,.3f}".format(desviacion_estandar))
```

### Rango

```python
salary_in_usd = df['salary_in_usd'].max() - df['salary_in_usd'].min()
formatted_salary_in_usd = f'${salary_in_usd:,.3f}'
print(f'Rango: {formatted_salary_in_usd}')
```

### Percentiles y rango intercuartílico

```python
print(f'Valor mínimo: ${df["salary_in_usd"].min():,.2f}')
print(f'Percentil 10: ${df["salary_in_usd"].quantile(0.1):,.2f}')
print(f'Percentil 25: ${df["salary_in_usd"].quantile(0.25):,.2f}')
print(f'Percentil 50: ${df["salary_in_usd"].median():,.2f}')
print(f'Percentil 75: ${df["salary_in_usd"].quantile(0.75):,.2f}')
print(f'Percentil 90: ${df["salary_in_usd"].quantile(0.9):,.2f}')
print(f'Valor máximo: ${df["salary_in_usd"].max():,.2f}')
```

```python
iqr = df["salary_in_usd"].quantile(0.75) - df["salary_in_usd"].quantile(0.25)
print(f'Rango Intercuartílico: ${iqr:,.2f}')
```

### Boxplot para `salary_in_usd`

```python
limite_bigotes = iqr * 1.5
print(f'Límite de los bigotes: ${limite_bigotes:,.2f}')
```

*   La caja, al estar delimitada por el percentil 25 y el percentil 75, indica que los datos se encuentran concentrados en valores entre `$105.700,00 USD` y `$186.000,00 USD. ` En este grupo se encuentran el 50% de nuestros datos.
*   El límite de los bigotes es `$121,534.50 USD`. Teniendo esto en cuenta, se puede concluir que los valores mayores a `$308.257,50 USD` son considerados **atípicos.**
*  El bigote inferior alcanzó a cubir el primer cuartil por completo. Podemos asumir que hay valores muy alejados del IQR y que se tiene una tendencia hacia los valores bajos del rango.

## Gráficas de densidad

### Análisis salarial según el entorno de trabajo

```python
df.groupby(['work_setting'])['work_setting'].count()
```

```python
by_salary_usd = df.groupby(level=0)[['salary_in_usd']].mean()
work_setting = df.groupby(level=0)['work_setting'].last()
merged = by_salary_usd.merge(work_setting, left_index=True, right_index=True)
hybrid = by_salary_usd[merged['work_setting'] == 'Hybrid']
in_person = by_salary_usd[merged['work_setting'] == 'In-person']
remote = by_salary_usd[merged['work_setting'] == 'Remote']
```

```python
# Configurar el estilo del gráfico
sns.set(style='whitegrid')
color_1 = sns.color_palette("flare")[2]
color_2 = sns.color_palette("flare")[3]
color_3 = sns.color_palette("flare")[5]

# Trazar la distribución KDE para tipo de trabajo híbrido
ax = sns.kdeplot(data=hybrid, x='salary_in_usd', fill=True, label='Híbrido', color=color_1)

# Trazar la distribución KDE para tipo de trabajo presencial en el mismo eje
sns.kdeplot(data=in_person, x='salary_in_usd', fill=True, ax=ax, label='Presencial', color=color_2)

# Trazar la distribución KDE para tipo de trabajo remoto en el mismo eje
sns.kdeplot(data=remote, x='salary_in_usd', fill=True, ax=ax, label='Remoto', color=color_3)

# Configurar título y etiquetas
ax.set_title('Distribuciones de los salarios según el entorno laboral', fontsize=13, pad=15)
ax.set(xlabel='Salario en USD')
ax.set(ylabel='Densidad')

# Mostrar leyenda
ax.legend(loc='upper right')
plt.show()
```

* ¿Cómo influye el entorno laboral en los salarios de los profesionales de ciencias de datos?

El entorno laboral parece no influir de manera significativa en los salarios; no obstante, el entorno laboral "Híbrido", por su Curtosis Alta, es el que tiene más tendencia a la media de `$150.000,00 USD` y mayor probabilidad de irse a valores extremos o atípicos (salarios más altos 😉).



### Análisis salarial según el tipo de contrato

```python
df.groupby(['employment_type'])['employment_type'].count()
```

```python
by_salary_usd = df.groupby(level=0)[['salary_in_usd']].mean()
employment_type = df.groupby(level=0)['employment_type'].last()
merged = by_salary_usd.merge(employment_type, left_index=True, right_index=True)
contract = by_salary_usd[merged['employment_type'] == 'Contract']
freelance = by_salary_usd[merged['employment_type'] == 'Freelance']
full_time = by_salary_usd[merged['employment_type'] == 'Full-time']
part_time = by_salary_usd[merged['employment_type'] == 'Part-time']
```

```python
# Configurar el estilo del gráfico
sns.set(style='whitegrid')
color_1 = sns.color_palette("flare")[0]
color_2 = sns.color_palette("flare")[5]
color_3 = sns.color_palette("flare")[2]
color_4 = sns.color_palette("flare")[3]

# Trazar la distribución KDE para Contract
ax = sns.kdeplot(data=contract, x='salary_in_usd', fill=True, label='Contrato', color=color_1)

# Trazar la distribución KDE para Freelance en el mismo eje
sns.kdeplot(data=freelance, x='salary_in_usd', fill=True, ax=ax, label='Freelance', color=color_2)

# Trazar la distribución KDE para Full-time en el mismo eje
sns.kdeplot(data=full_time, x='salary_in_usd', fill=True, ax=ax, label='Full-time', color=color_3)

# Trazar la distribución KDE para Part-time en el mismo eje
sns.kdeplot(data=part_time, x='salary_in_usd', fill=True, ax=ax, label='Part-time', color=color_4)

# Configurar título y etiquetas
ax.set_title('Distribuciones de los salarios según el tipo de contrato', fontsize=13, pad=15)
ax.set(xlabel='Salario en USD')
ax.set(ylabel='Densidad')

# Mostrar leyenda
ax.legend(loc='upper right')
plt.show()
```

* ¿Existe una diferencia salarial significativa entre empleados a tiempo completo y empleados a tiempo parcial o contratados?

La curtosis para el tipo de contrato "Freelance" es la más alta; sin embargo, sus datos se encuentran concentrados en un alrededor de `$70.000,00 USD.`

Por otra parte, el tipo de contrato "Contrato" tiene una curtosis baja. Los datos se distribuyen de manera más uniforme, pero es este tipo de contrato el que más extensión de colas tiene, yéndose hacia los valores atípicos por encima de los `$400.000,00 USD.`💸

Se intentará realizar Regresión Lineal con 2 variables categóricas: `job_category` y `experience_level`.

### `job_category`

```python
counts_job_category = df['job_category'].value_counts()
print(counts_job_category)
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# Codificación one-hot
df_dummies = pd.get_dummies(df['job_category'], drop_first=True)

# Añadir la variable numérica
X = pd.concat([df['salary_in_usd'], df_dummies], axis=1)

# Variable dependiente
y = df['salary_in_usd']

# Regresión lineal con statsmodels
X = sm.add_constant(X)  # Añadiendo una constante
model_sm = sm.OLS(y, X).fit()

# Obtener los coeficientes del modelo
coeficients = model_sm.params

# Coeficientes del modelo con statsmodels
coeficients_sm = model_sm.params
print(coeficients_sm)

# Construir la ecuación de regresión
equation = f"Salario = {coeficients['const']:.2f} + "

for feature, coef in zip(X.columns[1:], coeficients[1:]):
    equation += f"({coef:.2f} * {feature}) + "

# Eliminar el último signo más y espacio en blanco
equation = equation[:-2]

# Imprimir la ecuación
print(f"Ecuación de regresión: {equation}")
```

```python
# Codificación one-hot
df_dummies = pd.get_dummies(df['experience_level'], drop_first=True)

# Añadir la variable numérica
X = pd.concat([df['salary_in_usd'], df_dummies], axis=1)

# Variable dependiente
y = df['salary_in_usd']

# Regresión lineal con statsmodels
X = sm.add_constant(X)  # Añadiendo una constante
model_sm = sm.OLS(y, X).fit()

# Obtener los coeficientes del modelo
coeficients = model_sm.params

# Coeficientes del modelo con statsmodels
coeficients_sm = model_sm.params
print(coeficients_sm)

# Construir la ecuación de regresión
equation = f"Salario = {coeficients['const']:.2f} + "

for feature, coef in zip(X.columns[1:], coeficients[1:]):
    equation += f"({coef:.2f} * {feature}) + "

# Eliminar el último signo más y espacio en blanco
equation = equation[:-2]

# Imprimir la ecuación
print(f"Ecuación de regresión: {equation}")
```

Debido a la naturaleza de nuestro Dataset, no es posible predecir el comportamiento salarial con los atributos que se tienen actualmente.
